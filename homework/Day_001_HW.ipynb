```python
def mean_squared_error(y, yp):
    """
    請完成這個 Function 後往下執行
    """
    """
    計算 MSE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mse: MSE
    """
    mse = MSE = sum(pow((y - yp),2) ) / len(y)
    return mse

MSE = mean_squared_error(y, y_hat)
print("The Mean square error is %.3f" % (MSE))
```
